```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import pandas as pd
import librosa
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.layers import Dense
from tensorflow.keras import datasets, layers, models
import tensorflow as tf
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from tensorflow.keras.layers import Input, Conv2D, BatchNormalization, Dense
from tensorflow.keras.layers import ReLU, concatenate
from sklearn.linear_model import LogisticRegression
```

```python
path='/content/drive/MyDrive/assignment 1/annotations.csv'
relative_path='/content/drive/MyDrive/assignment 1/train/train/'
def load_data(path,relative_path):
    df=pd.read_csv(path)
    data=[]
    for filename in df['fname']:
        address=relative_path+filename
        data.append(np.load(address))
    label=df['label']
    return data,label
```

```python
data,label=load_data(path,relative_path)
```

```python
mx_len=0
mn_len=100000
tot=0
for spectrograms in data:
    mx_len=max(mx_len,spectrograms.shape[2])
    mn_len=min(mn_len,spectrograms.shape[2])
    tot+=spectrograms.shape[2]
tot,mn_len,mx_len
```

```python
def list_to_np(data,max_len):
    final_data=[]
    for spectrogram in data:
        x=spectrogram.flatten()
        y=np.zeros((max_len-spectrogram.shape[2])*64)
        y=np.append(y,x)
        z=np.zeros((max_len-spectrogram.shape[2])*64)
        y=np.append(y,z)
        final_data.append(y)
    return np.array(final_data)
```

```python
data=list_to_np(data,mx_len)
```

```python
train_x,test_x,train_y,test_y=train_test_split(data,label,test_size=0.2,random_state=28)
```

```python
log_reg=LogisticRegression(max_iter=30)
log_reg_clf=log_reg.fit(train_x,train_y)
```

```python
log_reg_clf.score(test_x,test_y)
```

```python
confusion_log_clf=confusion_matrix(log_reg_clf.predict(test_x),test_y)
confusion_log_clf
#f1_score(train_y_pred,train_y,average='weighted')
```
